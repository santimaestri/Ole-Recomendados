```python
import pandas as pd
pd.set_option("display.max_rows", None)
pd.set_option('display.max_colwidth', None)
```

```python
df = pd.read_csv('Offer_daily_2025-07-01.csv', na_values=['Does not apply', 'Other'])
df.head()
```

```python
#First chart with the top brands

top_brands = df.groupby('brand')['product'].nunique().sort_values(ascending=False).head(5).reset_index()
top_brands
```

```python
top_brands.to_csv('Topbrands.csv', index=False)
```

```python
#Second chart with the most discounted items
top_rated = df.groupby('product')[['discount_price_new', 'price_new', 'discount_percentage_new', 'url']].max().sort_values(by = 'discount_percentage_new', ascending=False).head(10).reset_index()

top_rated.to_csv('top_discounted.csv', index=False)
top_rated.reset_index()
```

```python
#Second chart with the best rated items
top_rated = (
    df.groupby('product')[['rating_quantity_new', 'rating_stars_new', 'discount_percentage_new', 'url']]
    .max()
    .query('rating_stars_new >= 4.5')
    .sort_values(by='rating_quantity_new', ascending=False)
    .head(10)
    .reset_index()
)
top_rated.to_csv('top_rated.csv', index=False)
top_rated.reset_index()
```

```python
filtered_df = df[df['product'].str.contains(r'notebook|laptop|portátil|netbook', case=False, na=False)]

our_pick = (
    filtered_df
    .groupby('product')[['rating_stars_new', 'rating_quantity_new', 'discount_percentage_new', 'discount_price_new', 'picture', 'url']]
    .max()
    .sort_values(by='discount_price_new', ascending=False)
    .head(10)
    .reset_index()
)

our_pick
```
